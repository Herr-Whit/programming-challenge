# Weather Data Challenge
In weather.csv youâ€™ll find the daily weather data of a single month. Read the file, then output the day number (column one Day) of the day with the smallest temperature spread (difference between maximum & minimum temperature of the day.) The maximum temperature is the second column MxT, the minimum temperature the third column MnT.

```python
import pandas as pd
```

Transform the file in the "Pandas" DataFrame format

```python
source_file = r'src\main\resources\de\exxcellent\challenge\weather.csv'
df = pd.read_csv(source_file)
```

```python
df
```

The dataset is quite small and visual inspection only shows a potential issue with the "1HrP TPcpn", which is 0 for all days. This is not relevant for our calculation, so we can proceed safely.

Calculate the difference in Temparature for each day.

```python
df['TDiff'] = df['MxT'] - df['MnT']
```

Check for each day, whether it is one of the days with the highest temparature difference.

```python
maximum_temparature = max(df['TDiff'])
```

```python
ismax = df['TDiff'] == maximum_temparature
```

Select datapoints, which reflect the highest temparature differences using boolean indexing. Isolate the "Day" entry. 

```python
df_with_highest_diff = df[ismax]
df_with_highest_diff
```

Note: I so happens that there is only one day with the highest temparature. I can confirm, by checking the table again (especially the 'TDiff' column).
